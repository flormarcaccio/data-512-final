### Importing libraries for cleaning and for analysis

```python
import pandas as pd
from matplotlib import pyplot as plt
from scipy import stats
import seaborn as sn
```

### Cleaning

For easier handling, I will reduce the dataset to the fields I will be using during the analysis. For a description of all variables please refer to [the documentation](data/EADA_2019.doc).

```python
df = df[['unitid', 'institution_name', 'city_txt', 'state_cd', 'zip_text', 'ClassificationCode', 'classification_name',
   'sector_cd', 'sector_name', 'EFMaleCount', 'EFFemaleCount', 'EFTotalCount', 'STUDENTAID_MEN', 'STUDENTAID_WOMEN',
   'STUDENTAID_TOTAL', 'IL_PARTIC_MEN', 'IL_PARTIC_WOMEN', 'RECRUITEXP_MEN', 'RECRUITEXP_WOMEN', 'IL_EXP_MEN',
   'IL_EXP_WOMEN', 'IL_REV_MEN', 'IL_REV_WOMEN', 'HDCOACH_SALARY_MEN', 'HDCOACH_SALARY_WOMEN', 
   'IL_MEN_FTHEADCOACH_MALE', 'IL_MEN_PTHEADCOACH_MALE','IL_MEN_FTHEADCOACH_FEM', 'IL_MEN_PTHEADCOACH_FEM',
   'IL_WOMEN_FTHDCOACH_MALE', 'IL_WOMEN_PTHDCOACH_MALE', 'IL_WOMEN_FTHDCOACH_FEM', 'IL_WOMEN_PTHDCOACH_FEM', 
   'IL_MEN_FTASCOACH_MALE', 'IL_MEN_PTASCOACH_MALE', 'IL_MEN_FTASSTCOACH_FEM', 'IL_MEN_PTASSTCOACH_FEM', 
   'IL_WOMEN_FTASCOACH_MALE', 'IL_WOMEN_PTASCOACH_MALE', 'IL_WOMEN_FTASTCOACH_FEM', 'IL_WOMEN_PTASTCOACH_FEM']]
df.columns = map(str.lower, df.columns)
df.shape
```

Assigning a zero to null values in numeric features.

```python
cols_updt = ['studentaid_men', 'studentaid_women', 'recruitexp_men' , 'recruitexp_women', 'hdcoach_salary_men',
             'il_men_ftheadcoach_male', 'il_men_ptheadcoach_male', 'il_men_ftheadcoach_fem', 'il_men_ptheadcoach_fem',
             'il_men_ftascoach_male', 'il_men_ptascoach_male', 'il_men_ftasstcoach_fem', 'il_men_ptasstcoach_fem',
             'il_women_ftascoach_male', 'il_women_ptascoach_male', 'il_women_ftastcoach_fem', 'il_women_ptastcoach_fem']
df.update(df[cols_updt].fillna(0))
```

Keeping institutions that satisfy specific criteria:

```python
#Institutions in the NCAA divisions
df = df[df.classificationcode.isin(list(range(1,8)))]
#Institutions that declared students
df = df[df.eftotalcount > 0]
#Institutions that declared participants in men and women's teams
df = df[df.il_partic_men > 0]
df = df[df.il_partic_women > 0]
#Institutions that received financial aid
df = df[df.studentaid_total > 0]
#Institutions from the US
df = df[~df.state_cd.isna()]

df.shape
```

Creating fields for the recruiting expenses, total revenue, total expenses and athletic financial aid per participant in men and women's teams, and a field with the difference between financial aid per participant by gender.

```python
df['recruitexp_participant_women'] = df['recruitexp_women']/df['il_partic_women']
df['recruitexp_participant_men'] = df['recruitexp_men']/df['il_partic_men']

df['rev_participant_women'] = df['il_rev_women']/df['il_partic_women']
df['rev_participant_men'] = df['il_rev_men']/df['il_partic_men']

df['exp_participant_women'] = df['il_exp_women']/df['il_partic_women']
df['exp_participant_men'] = df['il_exp_men']/df['il_partic_men']

df['aid_participant_women'] = df['studentaid_women']/df['il_partic_women']
df['aid_participant_men'] = df['studentaid_men']/df['il_partic_men']
df['diff_aid_participant'] = df['aid_participant_men']-df['aid_participant_women']
```

Creating a field for the difference in number of participants in men and women's teams.

```python
df['diff_participation'] = df['il_partic_men'] - df['il_partic_women']
```

Creating fields for the total number of head and assistant coaches in men and women's teams, total number of coaches (head+assistant), and the difference in total number of coaches between men's and women's teams.

```python
df['men_num_hdcoaches'] = df['il_men_ftheadcoach_male']+df['il_men_ptheadcoach_male']+df['il_men_ftheadcoach_fem']+df['il_men_ptheadcoach_fem']
df['women_num_hdcoaches'] = df['il_women_fthdcoach_male']+df['il_women_pthdcoach_male']+df['il_women_fthdcoach_fem']+df['il_women_pthdcoach_fem']
df['men_num_ascoaches'] = df['il_men_ftascoach_male']+df['il_men_ptascoach_male']+df['il_men_ftasstcoach_fem']+df['il_men_ptasstcoach_fem']
df['women_num_ascoaches'] = df['il_women_ftascoach_male']+df['il_women_ptascoach_male']+df['il_women_ftastcoach_fem']+df['il_women_ptastcoach_fem']

df['men_num_coaches'] = df['men_num_hdcoaches']+df['men_num_ascoaches']
df['women_num_coaches'] = df['women_num_hdcoaches']+df['women_num_ascoaches']

df['diff_num_coaches'] = df['men_num_coaches'] - df['women_num_coaches']
```

Students per coach in women and men's teams, and the difference between these two. 

```python
df['women_per_coach'] = df['il_partic_women']/df['women_num_coaches']
df['men_per_coach'] = df['il_partic_men']/df['men_num_coaches']
df['diff_students_per_coach'] = df['men_per_coach'] - df['women_per_coach']
```

Percentage of head and assistant coaches in men and women's teams by gender.

```python
df['men_perc_femhdcoaches'] = (df['il_men_ftheadcoach_fem']+df['il_men_ptheadcoach_fem'])/df['men_num_hdcoaches']
df['women_perc_malehdcoaches'] = (df['il_women_fthdcoach_male']+df['il_women_pthdcoach_male'])/df['women_num_hdcoaches']
df['men_perc_femascoaches'] = (df['il_men_ftasstcoach_fem']+df['il_men_ptasstcoach_fem'])/df['men_num_ascoaches']
df['women_perc_maleascoaches'] = (df['il_women_ftascoach_male']+df['il_women_ptascoach_male'])/df['women_num_ascoaches']
```

Creating a field with the difference between the mean salary of head coaches in men's teams and the mean salary of head coaches in women's teams.

```python
df['diff_salary'] = df['hdcoach_salary_men'] - df['hdcoach_salary_women']
```

### Creating reusable plotting functions

```python
def single_distribution_plot(data, title_graph, x_label, png_name):
    fig, ax = plt.subplots(figsize=(10,6))
    plt.hist(data, color='purple')
    plt.title(title_graph, fontsize=20)
    plt.xlabel(x_label, fontsize=15)
    plt.xticks(fontsize=15)
    plt.yticks(fontsize=15)
    plt.grid()
    plot_name = 'graphs/'+png_name+'.png'
    plt.savefig(plot_name , bbox_inches='tight')
    plt.show()

def dual_distribution_plot(data1, data2, title_graph, x_label, data1_label, data2_label, png_name):
    fig, ax = plt.subplots(figsize=(10,6))
    plt.hist(data1, density=True, histtype='barstacked', alpha=0.7, color='purple', label=data1_label)
    plt.hist(data2, density=True, histtype='barstacked', alpha=0.7, color='skyblue', label=data2_label)
    plt.legend(loc = 'upper right', fontsize=15)
    plt.title(title_graph, fontsize=20)
    plt.xlabel(x_label, fontsize=15)
    plt.xticks(fontsize=15)
    plt.yticks(fontsize=15)
    plt.grid()
    plot_name = 'graphs/'+png_name+'.png'
    plt.savefig(plot_name , bbox_inches='tight')
    plt.show()
```

### 1.2 Is there any difference between the amount of financial aid given to students in men's teams and the student aid for participant's in women's teams?  

Knowing that financial aid plays an important role on the financial variables, I would like to know if there is a difference in the amount of aid given to students in women's sports with the aid given to participants in men's teams.

To do this, I will be looking at the 'diff_aid_participant' variable (financial aid per student in men's team - financial aid per student in women's team). If an instutition treats students in men and women's teams equally, you would assume that this variable should be close to zero.

```python
df['diff_aid_participant'].describe()
```

From those statistics, we can see that the mean of the variable is quite close to zero, with a pretty high standard deviation. To complete this analysis, I will plot the distribution of the variable and perform a one-sample t-test under the null hypothesis that the mean difference is zero.

```python
data = df['diff_aid_participant']
title = "Difference of financial aid per participant by gender"
xlabel = "Financial aid difference"

single_distribution_plot(data, title, xlabel, 'financial_aid_difference_gender')
```

```python
stats.ttest_1samp(df['diff_aid_participant'],0)
```

Because the p-value is greater than 0.05, I found no evidence to say that the financial aid given to students in women's teams is different to the aid given to participants in men's teams.

### 1.3 Does having a men's football team affect the ratio of of financial aid given to women's teams?   

After reading the articles in [Background/Related Work](#background) and knowing the importance of football in US institutions, I would like to explore the financial aid distribution from the previous question between institutions that have football agains the ones that don't.

I will first separate the dataset in the two subsets using the classification code. The codes 3, 5 and 7 represent the institutions in NCAA Divisions I, II and III with no football.

```python
nofootball = df[df.classificationcode.isin([3,5,7])]['diff_aid_participant']
football = df[~df.classificationcode.isin([3,5,7])]['diff_aid_participant']
```

The variable I will be using for this analysis is diff_aid_participant, which the same as in question 1.2. I will plot the distribution of this variable for both subsets and perform a two-sample t-test under the null hypothesis that there is no difference between institutions that play football and the ones that don't.

```python
data1 = football
data2 = nofootball
title = "Difference in financial aid per participant between students in men and women's teams"
xlabel = "Financial aid difference"
data1_label = 'Football'
data2_label = 'No football'

dual_distribution_plot(data1, data2, title, xlabel, data1_label, data2_label, 'financial_aid_difference_football')
```

```python
stats.ttest_ind(nofootball, football)
```

In question 1.2 we found that there was no evidence to say that institution give a different amount of financial aid to participants in women's sports in comparison to the financial aid given to students in men's teams. However, because the p-value of the last test was lower than 0.05, we have evidence to say that institution with football have a higher mean difference in financial aid by gender than institutions that don't play football. This means than, on average, institutions that play football assign higher financial aid per participant to students in men's teams.  
This difference in behavior can clearly be seen on the graph.

### *Hypothesis 2: Participation in sports is different between men and women.*

When I talk about participation in sports, what I want to find out is if female students have the same opportunities as men in an institution. I assume opportunities are equal when an institution has an equal number of female and male students participating in sports. 

I will plot the distribution of participants in men and women's teams and then perform a t-test. For the test, I have created in the data processing stage a field called diff_participation with the difference of participants by gender (men-women). The t-test is under the null hypothesis that participation is equal, so the mean difference of participants by gender is zero.

```python
data1 = df['il_partic_women']
data2 = df['il_partic_men']
title = "Distribution of students participating in sports by gender"
xlabel = "Students in sports"
data1_label = "Female students participation"
data2_label = "Male students participation"

dual_distribution_plot(data1, data2, title, xlabel, data1_label, data2_label, 'participation_gender')
```

```python
stats.ttest_1samp(df['diff_participation'],0)
```

Because the p-value is less than 0.05, I have evidence to reject the null hypothesis and sat that participation in sports by gender is not equal. This means that institutions don't have the same number of athletic positions available for both men and women.
